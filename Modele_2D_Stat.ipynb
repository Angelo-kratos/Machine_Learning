# DROIT D'AJUSTEMENT

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[5,78,4,23,15,63,20,21,58,99,50,211,43,56,74,12,10,23,63,99]
y=[]
for index in range(0,20):
    y.append(((x[index]) * 15000))
#print(y)
#len(x)
```

# calcul moyenne

```python
total_x=0
total_y=0

for i in range(len(x)):
    total_x = total_x + x[i]   ##total des valeur de x
#print(total_x)   
for index in range(len(y)):
    total_y = y[index] + total_y
##print(total_y)

##MOYENNE
moyenne_x = total_x / len(x)
##print(moyenne_x)
moyenne_y = total_y / len(y)
##print(moyenne_y)
```

```python
##(x[i] - moyenne_x)
moyenne_yi = []
moyenne_xi = []
for indx in range(len(x)):
    moyenne_xi.append(x[indx] - moyenne_x) ####(x[i] - moyenne_x)
#print(moyenne_xi)

for indy in range(len(y)): 
    moyenne_yi.append(y[indy] - moyenne_y)   ####(y[i] - moyenne_y)
#print(moyenne_yi)

moyenne_yi_moyenne_xi =[]  ##(x[i] - X) (y[i] - Y)
for index_xy in range(len(x)):
    moyenne_yi_moyenne_xi.append(moyenne_xi[index_xy] * moyenne_yi[index_xy])
#print(moyenne_yi_moyenne_xi)

somme_xi_yi = 0    ##SOMME((x[i] - X) (y[i] - Y))
for t in range(len(moyenne_yi_moyenne_xi)):
    somme_xi_yi= somme_xi_yi + moyenne_yi_moyenne_xi[t]    
#print(somme_xi_yi) 

##((x[i]-X)**2)
moyenne_xi_carre = []
for index_xi in range(len(moyenne_xi)):
    moyenne_xi_carre.append(moyenne_xi[index_xi]**2)
#print(moyenne_xi_carre)

somme_xi_moyenne_x = 0
for index_xi_carre in range(len(moyenne_xi_carre)):
    somme_xi_moyenne_x = somme_xi_moyenne_x + moyenne_xi_carre[index_xi_carre]  ##SOMME((xi-X)**2)
   # print(somme_xi_moyenne_x)

    ##Y = aZ + b / Z=X
a = somme_xi_yi / somme_xi_moyenne_x
#print(a)

b = moyenne_y - (a * moyenne_x)
#print("b = ", b)

def t(z):
    return a*z + b

z = np.array([2,220])

plt.figure(figsize=(20,20))
plt.scatter(x,y)
plt.plot(z,t(z))
z = 56
plt.scatter(z,t(z),alpha=1,s=100)
```
